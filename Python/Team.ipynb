Team Analysis:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('../Cleaned_Dataset/cleaned_wpl_deliveries.csv')
```

```python
df.info()
```

```python
team_wickets = df[df["player_dismissed"].notna()].groupby("bowling_team")["player_dismissed"].count().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=team_wickets.values, y=team_wickets.index, palette="magma")
plt.xlabel("Total Wickets")
plt.ylabel("Teams")
plt.title("Total Wickets Taken by Each Team in WPL")
plt.show()
```

```python
team_runs = df.groupby("batting_team")["runs_of_bat"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=team_runs.values, y=team_runs.index, palette="viridis")
plt.xlabel("Total Runs")
plt.ylabel("Teams")
plt.title("Total Runs Scored by Each Team in WPL")
plt.show()
```

```python
team_runs = df.groupby("batting_team")["runs_of_bat"].sum()
team_balls = df.groupby("batting_team")["over"].count()
team_strike_rate = (team_runs / team_balls) * 100

team_strike_rate = team_strike_rate.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=team_strike_rate.values, y=team_strike_rate.index, palette="coolwarm")
plt.xlabel("Strike Rate")
plt.ylabel("Teams")
plt.title("Team with Best Batting Strike Rate in WPL")
plt.show()
```

```python
team_runs_conceded = df.groupby("bowling_team")["runs_of_bat"].sum()
team_overs_bowled = df.groupby("bowling_team")["over"].count() / 6  
team_economy = team_runs_conceded / team_overs_bowled

team_economy = team_economy.sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=team_economy.values, y=team_economy.index, palette="viridis")
plt.xlabel("Economy Rate")
plt.ylabel("Teams")
plt.title("Team with Best Bowling Economy in WPL")
plt.show()
```

```python
team_runs_conceded = df.groupby("bowling_team")["runs_of_bat"].sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=team_runs_conceded.values, y=team_runs_conceded.index, palette="Reds_r")
plt.xlabel("Total Runs Conceded")
plt.ylabel("Teams")
plt.title("Most Runs Conceded by Teams in WPL")
plt.show()
```
